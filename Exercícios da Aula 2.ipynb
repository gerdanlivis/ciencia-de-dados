```python
# 20) Mapeie terceira coluna (petallength) de tal forma que os valores numéricos sejam substituídos por valores textuais, seguindo de acordo com a tabela abaixo:

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = np.genfromtxt(url, delimiter=',', dtype='object')
nomes = ('sepallength', 'sepalwidth', 'petallength', 'petalwidth', 'species')

petallength = np.array(iris[:,2], dtype='float')

for i, item in enumerate(petallength):
    if (item < 3):
        iris[i][2] = "Pequena"
    elif (item >= 3 and item < 5):
        iris[i][2] = "Média"
    else:
        iris[i][2] = "Grande"
    
print(iris)
```

```python
# 21) Crie uma nova coluna chamada volume, preenchida com dados resultantes da seguinte equação:

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='object')
nomes = ('sepallength', 'sepalwidth', 'petallength', 'petalwidth', 'species')

petallength = np.array(iris_2d[:,2], dtype='float')
sepallength = np.array(iris_2d[:,2], dtype='float')
volume = []

for i, item in enumerate(petallength):
    volume.append((np.pi*item*sepallength[i]**2)/3)

iris_2d_volume = np.c_[np.array(iris_2d), np.array(volume)]
    
print(iris_2d_volume)

```
